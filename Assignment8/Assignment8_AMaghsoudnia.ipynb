```python
DF_weather = pd.read_csv(path_tempratureflie,sep=";",index_col=0)
DF_weather.head(24)
```

```python
oldIndex_weather=DF_weather.index
newParsedIndex_weather=pd.to_datetime(oldIndex_weather)
DF_weather.index=newParsedIndex_weather
DF_weather.columns
```

```python
DF_temperature=DF_weather[["temperature"]]
```

```python
DF_temperature.head()
```

```python
DF_irradianceSource=pd.read_csv(path_irradiancefile,sep=";",index_col=1)
DF_irradiance=DF_irradianceSource[["gen"]]
# Previous line could also be represented this way:
#DF_irradiance=DF_irradianceSource.loc[:,["gen"]]
DF_irradiance.head(24)
```

```python
DF_irradianceSource.columns
```

```python
DF_irradiance.loc[DF_irradiance.loc[:,"gen"]<0,:]=0
DF_irradiance.head(24)

```

```python
DF_joined = DF_consumption.join([DF_temperature,DF_irradiance])
```

```python
oldIndex_joined=DF_joined.index
newIndex_joined=pd.to_datetime(oldIndex_joined)
DF_joined.index=newIndex_joined
#DF_joined.loc[:,"temperature"].shift(-5)
DF_joined.head(8)
DF_joined_selected= DF_joined.loc["2014-07-01 00:00:00" : "2014-07-03 00:00:00"]
DF_joined_selected
```

```python
x = np.array(DF_joined_selected.index)
y1 = np.array(DF_joined_selected.iloc[:,0]) #consumption
y2 = np.array(DF_joined_selected.iloc[:,1]) #temperature
y3 = np.array(DF_joined_selected.iloc[:,2]) #gen
fig, axs = plt.subplots(3)
fig.suptitle('Vertically stacked subplots')
axs[0].plot(x, y1)
axs[0].set(xlabel='time', ylabel='AC Power (W)')
axs[1].plot(x, y2)
axs[1].set(xlabel='time', ylabel='Temperature (F)')
axs[2].plot(x, y3)
axs[2].set(xlabel='time', ylabel='Irradiance (W/m^2)')
```

```python
#normalized all in one
#df.loc[df['Value'].idxmax()] for locating
cmax=DF_joined_selected.loc[DF_joined_selected['air conditioner_5545'].idxmax()]
tmax=DF_joined_selected.loc[DF_joined_selected['temperature'].idxmax()]
imax=DF_joined_selected.loc[DF_joined_selected['gen'].idxmax()]

DF_joined_selected_normalized=DF_joined_selected
DF_joined_selected_normalized.loc[:,'air conditioner_5545']=DF_joined_selected_normalized.loc[:,'air conditioner_5545']/cmax['air conditioner_5545']
DF_joined_selected_normalized.loc[:,'temperature']=DF_joined_selected_normalized.loc[:,'temperature']/tmax['temperature']
DF_joined_selected_normalized.loc[:,'gen']=DF_joined_selected_normalized.loc[:,'gen']/imax['gen']

DF_joined_selected_normalized
```

```python
#normalized plot
x = np.array(DF_joined_selected_normalized.index)
y1 = np.array(DF_joined_selected_normalized.iloc[:,0]) #consumption
y2 = np.array(DF_joined_selected_normalized.iloc[:,1]) #temperature
y3 = np.array(DF_joined_selected_normalized.iloc[:,2]) #gen
fig, axs = plt.subplots(1)
fig.suptitle('Normalized')

plt.plot(x, y1, '-', label='AC Power (W)', color='b')
plt.plot(x, y2, '--', label='Temperature (F)', color='r')
plt.plot(x, y3, '-.', label='Irradiance (W/m^2)', color='y')
plt.legend(loc='best')
plt.show()
```
