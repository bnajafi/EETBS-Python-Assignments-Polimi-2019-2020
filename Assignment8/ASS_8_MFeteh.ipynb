```python
DF_weather = pd.read_csv(path_tempratureflie,sep=";",index_col=0)
DF_weather.head(24)
```

```python
DF_weather.index=pd.to_datetime(DF_weather.index)
newParsedIndex_weather=DF_weather.index
newParsedIndex_weather=newParsedIndex_weather.tz_localize("UTC").tz_convert('America/Mexico_City').tz_localize(None)
DF_weather.index=newParsedIndex_weather
DF_weather.columns
```

```python
DF_temperature=DF_weather[["temperature"]]
```

```python
DF_temperature.head()
```

```python
DF_irradianceSource=pd.read_csv(path_irradiancefile,sep=";",index_col=1)
DF_irradiance=DF_irradianceSource[["gen"]]
# Previous line could also be represented this way:
#DF_irradiance=DF_irradianceSource.loc[:,["gen"]]
DF_irradiance.head(24)
```

```python
DF_irradianceSource.columns
```

```python
DF_irradiance.loc[DF_irradiance.loc[:,"gen"]<0,:]=0
DF_irradiance.head(24)

```

```python
DF_consumption_First_week=DF_consumption.loc['2014-07-01 00:00:00':'2014-07-07 00:00:00',:]
DF_Temperature_First_week=DF_temperature.loc['2014-07-01 00:00:00':'2014-07-07 00:00:00',:]
DF_Irradiance_First_week=DF_irradiance.loc['2014-07-01 00:00:00':'2014-07-07 00:00:00',:]
%matplotlib notebook
plt.figure()
fig, ax = plt.subplots(3)
ax[0].plot(DF_consumption_First_week,"g")
ax[0].set(xlabel="time",ylabel="AC power")
ax[0].set_title("First week of July")
ax[1].plot(DF_Temperature_First_week,"b")
ax[1].set(xlabel="time",ylabel="Temp")
ax[1].set_title("First week of July")
ax[2].plot(DF_Irradiance_First_week,"r")
ax[2].set(xlabel="time",ylabel="Generation")
ax[2].set_title("First week of July")
```

```python
oldIndex_irradiance=DF_irradiance.index
newIndex_irradiance=pd.to_datetime(oldIndex_irradiance)
DF_irradiance.index=newIndex_irradiance
```
